```python
import networkx as nx
import sys
import random
sys.path.append('../src/')
from random_graph import RandGraph
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
g =RandGraph(moving=100, actors=10000, n_paths=100, path_depth=20, graph_type='hongo_district')
```

```python
# random actions
total_rewards = {}

for i in [100,70,30,0]:
    stored_rewards = []
    for j in range(30):
        action = random.sample(g.core_nodes, i)
        _, reward = g.action(action)
        stored_rewards.append(reward)
    total_rewards[i] = stored_rewards
```

```python
pd.DataFrame(total_rewards).plot(kind='box')
plt.show()
```

We always observe a better reward when no action is taken. The actors output is not the best reward to reduce congestion.

```python
# random actions
total_rewards = {}

for i in [100,70,30,0]:
    stored_rewards = []
    for j in range(30):
        action = random.sample(g.core_nodes, i)
        values, _ = g.action(action)
        stored_rewards.append(len(g.core_nodes)/np.sum(values))
    total_rewards[i] = stored_rewards
```

```python
pd.DataFrame(total_rewards).plot(kind='box')
plt.show()
```

```python
df = pd.DataFrame(total_rewards)
sns.swarmplot(data=df)
plt.show()
```
