```python
import networkx as nx
import random
import matplotlib.pyplot as plt
```

### Rational
We would like to design a graph with multiple entry and exit nodes.

```python
entry_nodes = [0,1,2,3,4]
exit_nodes = [5,6,7,8]
core_nodes = [(x,{'capacity': random.randint(1,10)}) for x in range(9,20)]
```

```python
g = nx.DiGraph()
```

```python
g.add_nodes_from(entry_nodes)
g.add_nodes_from(exit_nodes)
g.add_nodes_from(core_nodes)
```

```python
# paths = entry --> n x nodes --> exit
for i in range(20):
    edge_list = []
    # select random entry
    entry = random.choice(entry_nodes)

    # random path length
    path_len = random.randint(2,20)
    node1 = entry
    for edge in range(path_len):
        node2 = random.randint(9,20)
        edge_list.append((node1,node2))
        node1 = node2
    # add path to the graph
    g.add_edges_from(edge_list)
    # connect the last node to exit
    exit_node = random.choice(exit_nodes)
    g.add_edge(node1,exit_node)
```

```python
pos = nx.kamada_kawai_layout(g)
nx.draw_networkx_nodes(g, pos, node_color='steelblue')
nx.draw_networkx_labels(g,pos, font_color='w')
nx.draw_networkx_edges(g,pos)
plt.axis('off')
plt.show()
```
