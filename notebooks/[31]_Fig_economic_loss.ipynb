```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Traffic congestion economic loss: 

    NYC (33.7 M$, 8,6m pop, 1214 km2, 91 h/y),
LA (19.2 M$, 3.98m pop, 1302 km2, 102 h/y),
    SF (10.6M$, 0.884m pop, 600 km2, 79 h/y)

```python
economic_loss = [33.7, 19.2, 10.6]
population = [8.6, 3.98, 0.884]
surface = [1214, 1302, 600]
hour_lost = [91,102,79]
index = ['NYC','LA','SF']
col_names = ['economic_loss','population','hour_lost','surface']
```

```python
l = [economic_loss,population,hour_lost, surface]
data = np.array(l)
```

```python
data.shape
```

```python
df = pd.DataFrame(data.T, index=index, columns=col_names)
```

```python
df
```

```python
plt.figure(figsize=(8,8))
sns.set_context("paper", font_scale=1.8)
sns.set_style('darkgrid')
sns.set_palette('husl', n_colors=3)

plt.subplot(221)
sns.barplot(x=index, y=economic_loss, data=df)
plt.title('Estimated Economic loss \nof Traffic Congestion (M$)')

plt.subplot(222)
sns.barplot(x=index, y=population, data=df)
plt.title('Population (M)')

plt.subplot(223)
sns.barplot(x=index, y=hour_lost, data=df)
plt.title('Hour lost per year per driver')

plt.subplot(224)
sns.barplot(x=index, y=surface, data=df)
plt.title('Surface (kmÂ²)')

sns.despine()
plt.savefig('../fig/economic_loss.png', dpi=300)
plt.show()
```
