```python
from math import pi
```

```python
def get_nb_moving_actors(step_nb, max_moving=10):
    y = int((step_nb + pi) % (8 * pi))
    y /= np.std([0,max_moving])*4
    y *= max_moving 
    y = np.random.normal(loc=y, scale=max_moving/10, size=1)
    y = np.round(np.maximum(y,0))
    return int(y[0])
```

```python
get_nb_moving_actors(25)
```

```python
pd.DataFrame([get_nb_moving_actors(i) for i in range(100)]).plot(kind='line')
plt.show()
```
